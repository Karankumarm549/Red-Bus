```python
import pandas as pd
import mysql.connector
import numpy as npy
```

```python
dfb1=pd.read_csv('buses_AP.csv')
dfb2=pd.read_csv('buses_R.csv')
dfb3=pd.read_csv('buses_T.csv')
dfb4=pd.read_csv('buses_UP.csv')
dfb5=pd.read_csv('buses_G.csv')
dfb6=pd.read_csv('buses_WB.csv')
dfb7=pd.read_csv('buses_KL.csv')
dfb8=pd.read_csv('buses_H.csv')
dfb9=pd.read_csv('buses_SB.csv')
dfb10=pd.read_csv('buses_AS.csv')

Al_df=pd.concat([dfb1,dfb2,dfb3,dfb4,dfb5,dfb6,dfb7,dfb8,dfb9,dfb10],ignore_index=True)
```

```python
#convert prices to numeric
Al_df["Price"]=Al_df["Price"].str.replace("INR","")
Al_df["Price"]=Al_df["Price"].astype(float)
Al_df["Price"].fillna(0) 
```

```python

# Sample DataFrame for demonstration (you can replace this with your actual DataFrame)
#All_df = pd.DataFrame({"Ratings": ["New 5", "3", "New 2", None, "New 4", " ", "New 6", 7, 8.5]})

# Print original DataFrame
print("Original Ratings:")
print(Al_df["Ratings"])

# Step 1: Ensure all values in Ratings are converted to strings
Al_df["Ratings"] = Al_df["Ratings"].astype(str)
print("\nAfter converting to string:")
print(Al_df["Ratings"])

# Step 2: Replace "New" with an empty string
Al_df["Ratings"] = Al_df["Ratings"].str.replace("New", "", regex=False)
print("\nAfter replacing 'New':")
print(Al_df["Ratings"])

# Step 3: Strip whitespace from each element in the Series
Al_df["Ratings"] = Al_df["Ratings"].str.strip()
print("\nAfter stripping whitespace:")
print(Al_df["Ratings"])

# Step 4: Split and keep only the first part (if space-separated)
Al_df["Ratings"] = Al_df["Ratings"].str.split().str[0]
print("\nAfter splitting and keeping first part:")
print(Al_df["Ratings"])

# Step 5: Convert to numeric, coerce errors to NaN
Al_df["Ratings"] = pd.to_numeric(Al_df["Ratings"], errors='coerce')
print("\nAfter converting to numeric:")
print(Al_df["Ratings"])

# Step 6: Fill NaN values with 0
Al_df["Ratings"].fillna(0, inplace=True)
print("\nAfter filling NaN values with 0:")
print(Al_df["Ratings"])

# Final DataFrame
print("\nFinal DataFrame:")
print(Al_df)
```
